# Linear Regression applying Normal Equation

```python
import pandas as pd
import numpy as np
import scipy.linalg as sp
import matplotlib.pyplot as plt
#import the dataset

df = pd.read_csv("house_prices_multivariate.csv")
x_matrix = df.iloc[:,:-1].values #take all columns expect the last column
y_vector = df.iloc[:,-1].values #take the last column
x_trans = x_matrix.T #transpose of the above matrix
x_input = x_trans.dot(x_matrix) #x transpose times X
x_part1 = sp.inv(x_input) #inverse of the above matrix
x_part2 = x_trans.dot(y_vector) # x_transpose times Y
parameter_values = x_part1.dot(x_part2) #apply formula  (XT*X)-1 * (XT*y)
y_pred = []
for i in range(len(x_matrix)):
    y_pred.append(sum(x_matrix[i]*parameter_values))
#difference between actual and predicted values    
diff = y_pred - y_vector
print(diff)
plt.plot(diff)
plt.show()
```

```python
from sklearn import preprocessing
df = pd.read_csv("house_prices_multivariate.csv")
x_matrix = df.iloc[:,:-1].values #take all columns expect the last column
y_vector = df.iloc[:,-1].values #take the last column
final_matrix = preprocessing.scale(x_matrix)
x_trans = final_matrix.T #transpose of the above matrix
x_input = x_trans.dot(final_matrix) #x transpose times X
x_part1 = np.linalg.inv(x_input) #inverse of the above matrix
x_part2 = x_trans.dot(y_vector) # x_transpose times Y
parameter_values = x_part1.dot(x_part2) #apply formula  (XT*X)-1 * (XT*y)
y_pred = []
for i in range(len(final_matrix)):
    y_pred.append(sum(final_matrix[i]*parameter_values))
#difference between actual and predicted values    
diff = y_pred - y_vector
print(diff)
plt.plot(diff)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression    
lr =LinearRegression()
my_model = lr.fit(x_matrix, y_vector)
plt.plot(my_model.predict(x_matrix)-y_vector)
plt.show()
```
