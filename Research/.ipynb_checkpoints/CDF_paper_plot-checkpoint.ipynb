```python
from mpl_toolkits.axes_grid1 import host_subplot
import mpl_toolkits.axisartist as AA
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def ECR(n):
    return ((-100 + n)*(-1 + math.sqrt(n/(18+n))) + n*(1+1/math.sqrt((18+n)/n)))/math.sqrt(math.pi)
```

```python
def prob(n):
    return 1- ((1/2) + math.atan(math.sqrt(n)/(3*math.sqrt(2)))/(math.pi))
```

```python
def sample(n):
    return n/2
```

```python
vecfunc = np.vectorize(ECR)
vecfunc1 = np.vectorize(prob)
vecfunc2 = np.vectorize(sample)
d = np.arange(0.1, 100.0, 0.01)
T = vecfunc(d)
T1 = vecfunc1(d)
T2 = vecfunc2(d)

T_hat = np.vstack((T,T1,T2))
print(T_hat[0])
```

```python
host = host_subplot(111, axes_class=AA.Axes)
plt.subplots_adjust(right=0.75)

par1 = host.twinx()
par2 = host.twinx()

offset = 60
new_fixed_axis = par2.get_grid_helper().new_fixed_axis
par2.axis["right"] = new_fixed_axis(loc="right", axes=par2,
                                        offset=(offset, 0))

par2.axis["left"].toggle(all=False)

host.set_xlim(0, 100)
par1.set_ylim(0,1)
par2.set_ylim(0,100)

host.set_xlabel("Total number of experimental units (2n)", fontsize=16)
host.set_ylabel("Expected Cumulative Regret")
par1.set_ylabel("Prob of choosing the sub optimal treatment (1-P(L*))")
par2.set_ylabel("Sub optimal units subject to experimentation(n)")

p1, = host.plot(d, T, label="Expected Cumulative Regret")
p2, = par1.plot(d, T1, label="Probability of choosing the sub optimal treatment")
p3, = par2.plot(d,T2, label="Sub optimal experiments")

host.legend()
host.axis["right"].label.set_color(p1.get_color())
par1.axis["left"].label.set_color(p2.get_color())
par2.axis["left"].label.set_color(p3.get_color())

par1.spines["left"].set_position(("axes", -0.4)) # red one
par2.spines["left"].set_position(("axes", -0.2)) # green one

par1.spines["left"].set_visible(True)
par1.yaxis.set_label_position('left')
par1.yaxis.set_ticks_position('left')

par2.spines["left"].set_visible(True)
par2.yaxis.set_label_position('left')
par2.yaxis.set_ticks_position('left')


host.legend(loc=2, prop={'size': 7})
plt.draw()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# To make things reproducible...
np.random.seed(1977)

fig, ax = plt.subplots()

# Twin the x-axis twice to make independent y-axes.
axes = [ax, ax.twinx(), ax.twinx()]

# Make some space on the right side for the extra y-axis.
fig.subplots_adjust(right=0.75)

# Move the last y-axis spine over to the right by 20% of the width of the axes
axes[1].spines['right'].set_position(('axes', -0.25))
axes[2].spines['right'].set_position(('axes', -0.5))

# To make the border of the right-most axis visible, we need to turn the frame
# on. This hides the other plots, however, so we need to turn its fill off.
axes[-1].set_frame_on(True)
axes[-1].patch.set_visible(False)

# And finally we get to plot things...
colors = ('Green', 'Red', 'Blue')
intAxNo = 0
for ax, color in zip(axes, colors):
    intAxNo += 1
    ax.plot(d,T_hat[intAxNo-1], color=color)
    if (intAxNo > 1):
        if (intAxNo == 2):
            ax.set_ylabel('Prob of choosing the sub optimal treatment (1-P(L*))' , color=color, labelpad = -40 )
        elif (intAxNo == 3):
            ax.set_ylabel('Sub optimal units subject to experimentation(n)', color=color, labelpad = -45 )
        ax.get_yaxis().set_tick_params(direction='out')
    else:
        ax.set_ylabel('Expected Cumulative Regret', color=color, labelpad = +0 )

    ax.tick_params(axis='y', colors=color)
axes[0].set_xlabel('Total number of experimental units (2n)')
axes[0].set_xlim(0, 100)
axes[1].set_ylim(0,1)
axes[2].set_ylim(0,100)
axes[-3].set_ylim(-40,90)



plt.savefig('figurenew2.png',bbox_inches='tight',dpi=300)

```
