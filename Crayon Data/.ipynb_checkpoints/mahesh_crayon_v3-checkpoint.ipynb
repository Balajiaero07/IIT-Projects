```python
import re

def search(text,target,n):
    '''Searches for text, and retrieves n words either side of the text, which are retuned seperatly'''
    word = r"\W*([\w]+)"
    groups = re.search(r'{}\W*{}{}'.format(word*n,target,word*n), text).groups()
    aspect = list(groups)
    #return groups[:n],groups[n:]
    return aspect
```
