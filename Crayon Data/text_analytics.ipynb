```python
os.chdir("E:\\Consultancy\\Crayon")
df = DataFrame.from_csv("100K_Restaurants_reviews.tsv", sep="\t")
```

```python
count_vect = CountVectorizer()
count_data = count_vect.fit_transform(df['ReviewTxt'])
count_vect.vocabulary_.get(u'algorithm')
```

```python
len(df)
```

```python
stemmed_words = []
for i in range(len(df)):
    words = word_tokenize(df.iloc[i,9])
    for w in words:
        stemmed_words.append(ps.stem(w))
```

```python
c = Counter(stemmed_words)
for i,j in c.most_common():
    print (j,i)
```

```python
tf_transformer = TfidfTransformer(use_idf=False).fit(count_data)
tf_data = tf_transformer.transform(count_data)
tf_data.shape
```

```python
stop = stopwords.words('english')
df['without_stopwords'] = df['ReviewTxt'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
df['without_stopwords'].replace(to_replace='The', value="",regex=True)
df['without_stopwords'].replace(to_replace='the', value="",regex=True)
df['without_stopwords'].replace(to_replace='I', value="",regex=True)
df['without_stopwords'].replace(to_replace='i', value="",regex=True)
df['without_stopwords'].replace(to_replace='this', value="",regex=True)
df['without_stopwords'].str.replace('[{}]'.format(string.punctuation), '')
```

```python
df['POS'] = pos_tag_sents(df['without_stopwords'].apply(word_tokenize).tolist())
```

```python
noun_words = []
for j in range(len(df)):
    tagged = df.iloc[j,10]
    for item in tagged:
        if item[1][0] == 'N' or item[1][0] == 'NN' or item[1][0] == 'NNP':
            noun_words.append(item[0])
```

```python
c = Counter(noun_words)
for i,j in c.most_common():
    print (j,i)
```
